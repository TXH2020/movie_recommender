```python
import gdown
import json
import requests
import time
url1="https://drive.google.com/uc?id=1kG9Ajf7xYDuuCj-bUPWggh9Tc3SYOgKa"
gdown.download(url1,"movies.json", quiet=True)
```

```python
actual_movies=[]
def save_as_csv():
    movies=None
    with open('movies.json','r') as f:
      movies=f.readlines()
    for i in movies:
      actual_movies.append(json.loads(i.rstrip('\n')))

save_as_csv()
```

```python
films = []
file = open('data.json','w')
url = """https://api.themoviedb.org/3/movie/{}?api_key=48b55ffa37b6a2627bca2161a21e62bb"""
count=1
for page in actual_movies:
      if(count==300000):
        break
      try:
        r = requests.get(url.format(page['id']))
        film=r.json()
        f = dict()
        f['id'] = page['id']
        f['imdb_id'] = film['imdb_id']
        f['title'] = film['original_title']
        f['description'] = film['overview']
        genres=[]
        for i in film['genres']:
          genres.append(i['id'])
        f['genres'] = genres
        films.append(f)
        print(count)
        count+=1
      except Exception as e:
        print(e)

json.dump(films, file, sort_keys=True, indent=4)

file.close()
```

```python
import json
import pandas as pd
database=[]
def get_descriptions():
    #MovieDescriptions.objects.all().delete()
    data=None
    with open('new_desc.json','r') as f:
        data=json.load(f)
    df=pd.read_csv('desc.csv')
    for i in range(df.shape[0]):
            film=df.iloc[i]
            f={}
            f['id'] = film['id']
            try:
                f['imdb_id'] = str(int(film['imdb_id']))
            except:
                f['imdb_id'] = film['imdb_id']
            f['title'] = film['title']
            f['description'] = film['description']
            f['genres'] = film['genres']
            database.append(f)
            
    for film in data:
            f={}
            f['id'] = film['id']
            try:
                f['imdb_id'] = str(film['imdb_id'])[2:]
            except:
                f['imdb_id'] = film['imdb_id']
            f['title'] = film['title']
            f['description'] = film['description']
            f['genres'] = film['genres']
            database.append(f)
```
