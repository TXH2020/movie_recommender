```python
import pandas as pd
movies_df=pd.read_csv("https://raw.githubusercontent.com/TXH2020/movie_recommender/main/csv_files/movgmov.csv")
```

```python
mov_dict=dict()
for i in range(movies_df.shape[0]):
  mov_dict[movies_df.iloc[i][0]] = movies_df.iloc[i][1]
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/TXH2020/movie_recommender/main/csv_files/rating.csv")
d=dict(df['movie_id'].value_counts())
```

```python
import numpy as np
c=0
lim=101
ar=np.arange(0,lim)
for i in d.keys():
  if(d[i] in ar):
    c+=1
print(c)
```

```python
movids=[]
for i in x:
  movids.append(int(i[0]))
```

```python
c=0
match_movies=[]
for i in movids:
  if i in d.keys():
    print(d[i])
```

```python
mdf=pd.DataFrame({'movid':d.keys(),'count':d.values()})
```

```python
rel=mdf.nlargest(3997,'count')
```

```python
rel['mov_names']=rel['movid'].apply(lambda x: mov_dict[x])
```

```python
def transform(i):
  i = i.split(' ')
  i = '+'.join(i)
  i = i.rstrip('+')
  return i
rel['mov_names']=rel['mov_names'].apply(transform)
```

```python
import pickle
with open('movname.pkl','wb') as f:
  pickle.dump(list(rel['mov_names']),f)
```
